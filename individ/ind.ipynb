## Вычислить площадь между двумя заданными кривыми на определенном интервале. Затем постройте график этих кривых и закрашенной области между ними.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

## Задаём функции двух кривых.

```python
def f(x):
    return x**2

def g(x):
    return np.sin(x)
```

```python
x = np.linspace(0, 2*np.pi, 100) #интервал x
```

```python
# вычисляем площадь между кривыми 
area, error = integrate.quad(lambda x: f(x) - g(x), 0, 2*np.pi)  # Вычисление интеграла
print("Площадь между кривыми:", area)
```

```python
plt.plot(x, f(x), label='f(x)')
plt.plot(x, g(x), label='g(x)')
plt.fill_between(x, f(x), g(x), where=(f(x) > g(x)), color='gray', alpha=0.5)
plt.text(1, 2, "Площадь = {:.2f}".format(area), fontsize=12)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
